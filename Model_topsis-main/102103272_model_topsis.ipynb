```python

from datasets import load_dataset
dataset = load_dataset("tweet_eval",'sentiment',split='test')
```

```python
dataset
```

```python
model_path="cardiffnlp/twitter-roberta-base-sentiment-latest"
from transformers import pipeline
sentiment_task = pipeline("sentiment-analysis", model=model_path, tokenizer=model_path)
```

```python
test=dataset[0:1000]['text']
```

```python
test
```

```python
output=sentiment_task(test)
```

```python
label_mapping = {'positive': 2, 'neutral': 1,'negative':0}
predicted_labels_ = [label_mapping[result['label']] for result in output]
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,matthews_corrcoef,confusion_matrix,roc_auc_score
accuracy1 = accuracy_score(dataset[0:1000]['label'], predicted_labels_)
precision1 = precision_score(dataset[0:1000]['label'], predicted_labels_, average='weighted')
recall1 = recall_score(dataset[0:1000]['label'], predicted_labels_, average='weighted')
f1_1 = f1_score(dataset[0:1000]['label'], predicted_labels_, average='weighted')
```

```python
print(f"Accuracy: {accuracy1}")
print(f"Precision: {precision1}")
print(f"Recall: {recall1}")
print(f"F1 Score: {f1_1}")
```

```python
model_path="austinmw/distilbert-base-uncased-finetuned-tweets-sentiment"
from transformers import pipeline
sentiment_task = pipeline("sentiment-analysis", model=model_path, tokenizer=model_path)
```

```python
output=sentiment_task(test)
```

```python
label_mapping = {'LABEL_2': 2, 'LABEL_1': 1,'LABEL_0':0}
predicted_labels_ = [label_mapping[result['label']] for result in output]
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,matthews_corrcoef,confusion_matrix,roc_auc_score
accuracy2 = accuracy_score(dataset[0:1000]['label'], predicted_labels_)
precision2 = precision_score(dataset[0:1000]['label'], predicted_labels_, average='weighted')
recall2 = recall_score(dataset[0:1000]['label'], predicted_labels_, average='weighted')
f1_2 = f1_score(dataset[0:1000]['label'], predicted_labels_, average='weighted')
```

```python
print(f"Accuracy: {accuracy2}")
print(f"Precision: {precision2}")
print(f"Recall: {recall2}")
print(f"F1 Score: {f1_2}")
```

```python
import pandas as pd
data = {
    'Model': ['RoBERTa', 'DistilBERT'],
    'Accuracy': [accuracy1, accuracy2],
    'Precision': [precision1, precision2],
    'Recall': [recall1, recall2],
    'F1 Score': [f1_1, f1_2]
}

df = pd.DataFrame(data)
```

```python
df.to_csv('result.csv')
```
